```python
from transformers import pipeline
from docx import Document
from google.colab import files

# Load a larger pre-trained model for question answering
qa_pipeline = pipeline('question-answering', model='bert-large-uncased-whole-word-masking-finetuned-squad')

# Read the .docx file
docx_file = 'sample_policy_document.docx'

document = Document(docx_file)
doc_text = "\n".join([paragraph.text for paragraph in document.paragraphs])

# Lists to store asked questions and their corresponding answers
asked_questions = []
given_answers = []

# Get user input for questions
while True:
    question = input("Ask a question (or type 'exit' to quit): ")

    if question.lower() == 'exit':
        break

    # Use the QA model to find the answer, adjusting topk and topk_per_candidate for more answers
    answer = qa_pipeline(question=question, context=doc_text, topk=3, topk_per_candidate=3)

    # Find the answer with the highest confidence
    best_answer = max(answer, key=lambda x: x['score'])

    # Store the asked question and the corresponding answer
    asked_questions.append(question)
    given_answers.append(best_answer['answer'])

    print("Best Answer:")
    print(f"- '{best_answer['answer']}' with confidence {best_answer['score']}")

# Generate a new document summarizing asked questions and given answers
output_doc = Document()
output_doc.add_heading('QA Summary', level=1)

for i, question in enumerate(asked_questions):
    output_doc.add_paragraph(f"Q: {question}")
    output_doc.add_paragraph(f"A: {given_answers[i]}")
    output_doc.add_paragraph("----------------------")

# Save the generated document
output_filename = 'QA_Summary.docx'
output_doc.save(output_filename)

# Provide a download link for the generated document
files.download(output_filename)
```
